```python
IPO2022 = pd.read_csv("IPO/ipo.csv")
IPO2022.head()
```

```python
IPO22_tickers = ""

for t in IPO2022["Symbol"]:
    IPO22_tickers += " " + str(t)
    
IPO22_tickers = IPO22_tickers[1:]
IPO22_tickers = set(IPO22_tickers.split())
IPO22_tickers
```

```python
# import IPO 2022 stock data
# Constants

start_date = '2019-01-01'
end_date = '2023-06-01'

IPOdata = yf.download(IPO22_tickers, start_date, end_date)

IPOdata.head()
```

```python
#  2019 !

IPOdata_sorted = IPOdata.dropna(axis=1)
IPOdata_sorted.head()
```

```python
# helper fucntions 

def get_ticker(tickers_df:pd.DataFrame, ticker, cols) -> pd.DataFrame:
    
    # retrieve the ticker's following info
    new_cols = [(c, ticker) for c in cols]
    df = tickers_df[new_cols]
    
    # drop multiindex 
    df.columns = df.columns.droplevel(1)
    
    print(ticker, "dataframe has been created!")
    return df

def get_trend(ticker_df:pd.DataFrame) -> None:
        
    for col in ticker_df.columns:
        ticker_df[col].plot()
        plt.title(col)
        plt.show()
```

```python
cols = ["Adj Close", "Close", "High", "Low", "Open", "Volume"]

ABST = get_ticker(IPOdata_sorted,"ABST", cols)
ABST.head()
```

```python
get_trend(ABST)
```

```python
start_date = pd.Timestamp(2019,1,2)
after_1year = pd.Timestamp(2019,12,31)
after_2year = pd.Timestamp(2020,12,31)
after_3year = pd.Timestamp(2021,12,31)

IPOdata_sorted.head()
```

```python
initial = IPOdata_sorted[IPOdata_sorted.index == start_date]
y1 = IPOdata_sorted[IPOdata_sorted.index == after_1year]
y2 = IPOdata_sorted[IPOdata_sorted.index == after_2year]
y3 = IPOdata_sorted[IPOdata_sorted.index == after_3year]

yearly_earning = pd.concat([initial,y1,y2,y3])
yearly_earning.head()
```

```python
cols = ["Adj Close"]

tickers = yearly_earning.columns.droplevel(0) # get tickers

analyzing_dic = {
    "ticker" : [],
    "start" : [],
    "y1after" : [],
    "y2after" : [],
    "y3after" : [],
}

for ti in tickers:
    analyzing_dic["ticker"].append(ti)
    analyzing_dic["start"].append(initial.at[start_date,("Adj Close", ti)])
    analyzing_dic["y1after"].append(y1.at[after_1year,("Adj Close", ti)])
    analyzing_dic["y2after"].append(y2.at[after_2year,("Adj Close", ti)])
    analyzing_dic["y3after"].append(y3.at[after_3year,("Adj Close", ti)])
```

```python
trend = pd.DataFrame(analyzing_dic)
trend.head()
```

```python
trend["y1_ratio"] = trend["y1after"] / trend["start"]
trend["y2_ratio"] = trend["y2after"] / trend["start"]
trend["y3_ratio"] = trend["y3after"] / trend["start"]

trend.head()
```
