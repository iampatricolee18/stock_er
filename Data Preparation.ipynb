```python
IPO2022 = pd.read_csv("IPO_2022.csv")
IPO2022.head()
```

```python
IPO22_tickers = ""

for t in IPO2022["Symbol"]:
    IPO22_tickers += " " + str(t)
    
IPO22_tickers = IPO22_tickers[1:]
IPO22_tickers
```

```python
# import IPO 2022 stock data
# Constants

start_date = '2023-01-01'
end_date = '2023-06-01'

IPO22data = yf.download(IPO22_tickers, start_date, end_date)

IPO22data.head()
```

```python
# helper fucntions 

def get_ticker(tickers_df:pd.DataFrame, ticker, cols) -> pd.DataFrame:
    
    # retrieve the ticker's following info
    new_cols = [(c, ticker) for c in cols]
    df = tickers_df[new_cols]
    
    # drop multiindex 
    df.columns = df.columns.droplevel(1)
    
    print(ticker, "dataframe has been created!")
    return df

def get_trend(ticker_df:pd.DataFrame) -> None:
        
    for col in ticker_df.columns:
        print(col)
        ticker_df[col].plot()
        plt.show()
```

```python
cols = ["Adj Close", "Close", "High", "Low", "Open", "Volume"]

get_ticker(IPO22data,"ACAB", cols)
```

```python
get_trend(ACAB)
```
