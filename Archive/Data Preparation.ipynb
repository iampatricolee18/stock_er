```python
# helper fucntions 

def get_ticker(tickers_df:pd.DataFrame, ticker, cols) -> pd.DataFrame:
    
    # retrieve the ticker's following info
    new_cols = [(c, ticker) for c in cols]
    df = tickers_df[new_cols]
    
    # drop multiindex 
    df.columns = df.columns.droplevel(1)
    
    print(ticker, "dataframe has been created!")
    return df

def get_trend(ticker_df:pd.DataFrame) -> None:
        
    for col in ticker_df.columns:
        ticker_df[col].plot()
        plt.title(col)
        plt.show()
```

```python
import calendar
import datetime

def get_last_friday_of_month(year, month):
    last_day = calendar.monthrange(year, month)[1]  # Get the last day of the month
    last_date = datetime.date(year, month, last_day)

    # Find the last Friday by checking previous days
    while last_date.weekday() != calendar.FRIDAY:
        last_date -= datetime.timedelta(days=1)
    
    d = last_date.day
    m = last_date.month
    y = last_date.year 
    
    return pd.Timestamp(y,m,d)

# Get the current year and month
current_date = datetime.date.today()
current_year = current_date.year
current_month = current_date.month

# Get the last Friday of the current month

for i in range(1,13):

    last_friday = get_last_friday_of_month(current_year, i)

    print("Last Friday of the month:", last_friday)
    #print(type(last_friday))
```
